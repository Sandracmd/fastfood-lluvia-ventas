```python
lluvia_region = lluvia_region.rename(columns={"valor": "lluvia_promedio"})
```

```python
ventas_clima = ventas_ticket.merge(lluvia_region, how="left", on=["region_id", "fecha_redondeada"])
```

```python
ventas_agrupadas = ventas_clima.groupby(["region_id", "fecha_redondeada"]).size().reset_index(name="ventas_por_hora")
```

```python
df_modelo = ventas_agrupadas.merge(lluvia_region, on=["region_id", "fecha_redondeada"])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
X = df_modelo[["lluvia_promedio"]]
y = df_modelo["ventas_por_hora"]
```

```python
modelo_lineal = LinearRegression()
modelo_lineal.fit(X, y)
df_modelo["pred_lineal"] = modelo_lineal.predict(X)
r2_lineal = r2_score(y, df_modelo["pred_lineal"])
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
modelo_poly = LinearRegression().fit(X_poly, y)
df_modelo["pred_poli"] = modelo_poly.predict(X_poly)
r2_poly = r2_score(y, df_modelo["pred_poli"])
```

```python
modelo_rf = RandomForestRegressor(n_estimators=100, random_state=42)
modelo_rf.fit(X, y)
df_modelo["pred_rf"] = modelo_rf.predict(X)
r2_rf = r2_score(y, df_modelo["pred_rf"])
```

```python
plt.figure(figsize=(16, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x=X["lluvia_promedio"], y=y, alpha=0.5)
sns.lineplot(x=X["lluvia_promedio"], y=df_modelo["pred_lineal"], color='red')
plt.title(f"Regresión Lineal\nR² = {r2_lineal:.4f}")
```

```python
plt.subplot(1, 3, 2)
sns.scatterplot(x=X["lluvia_promedio"], y=y, alpha=0.5)
sns.lineplot(x=X["lluvia_promedio"], y=df_modelo["pred_poli"], color='green')
plt.title(f"Regresión Polinomial\nR² = {r2_poly:.4f}")
```

```python
plt.subplot(1, 3, 3)
sns.scatterplot(x=X["lluvia_promedio"], y=y, alpha=0.5)
sns.lineplot(x=X["lluvia_promedio"], y=df_modelo["pred_rf"], color='orange')
plt.title(f"Random Forest\nR² = {r2_rf:.4f}")
plt.tight_layout()
plt.show()
```
